```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline

# we will limit the number of decimal places we display
pd.options.display.float_format = '{:,.2f}'.format 
```

```python
datafile_name= "C:/Users/lenovo/Downloads/Product demand project (6)/Product demand project/Data/Data/Customer_Details.xlsx"
abt_customer = pd.read_excel(datafile_name, index_col=0)

abt_customer.head(3)
```

```python
datafile_name= "C:/Users/lenovo/Downloads/Product demand project (6)/Product demand project/Data/Data/Order_Details.xlsx"

abt_order = pd.read_excel(datafile_name, index_col=0)

abt_order.head(10)
```

```python
datafile_name= "C:/Users/lenovo/Downloads/Product demand project (6)/Product demand project/Data/Data/Product_Details.xlsx"
abt_product = pd.read_excel(datafile_name, index_col=0)

abt_product.head(10)
```

```python
print(abt_customer.columns)
print("\n Dataset has %d rows and %d columns" % abt_customer.shape)
```

```python
print(abt_order.columns)
print("\n Dataset has %d rows and %d columns" % abt_order.shape)
```

```python
print(abt_product.columns)
print("\n Dataset has %d rows and %d columns" % abt_product.shape)
```

```python
abt_customer_order = pd.merge(abt_customer, abt_order, on="Row ID", how="outer")

# examine the combined ABT
display(abt_customer_order)
print(abt_customer.shape)
print(abt_order.shape)
print(abt_customer_order.shape)

print("\n Dataset has %d rows and %d columns" % abt_customer_order.shape)
```

```python
abt_final = pd.merge(abt_customer_order, abt_product, on="Row ID", how="outer")

# examine the combined ABT
display(abt_final)
print(abt_customer_order.shape)
print(abt_product.shape)
print(abt_final.shape)

print("\n Dataset has %d rows and %d columns" % abt_final.shape)
```
