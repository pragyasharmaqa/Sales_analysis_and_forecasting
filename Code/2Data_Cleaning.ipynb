```python
#fill the missing value in the column country

abt["Country"]=abt["Country"].fillna("United States")
```

```python
#fill the missing value in the column Postal Code

city_to_zip = dict()
for i in abt.index:
    if pd.notna(abt.loc[i, "Postal Code"]):
        city_to_zip[abt.loc[i, "City"]] = abt.loc[i, "Postal Code"]
for i in abt.index:
    if pd.isna(abt.loc[i, "Postal Code"]):
        abt.loc[i, "Postal Code"] = city_to_zip.get(abt.loc[i, "City"])
display(abt)
```

```python
#fill the missing value in the column Top N Customer Label

city_to_zip = dict()
for i in abt.index:
    if pd.notna(abt.loc[i, "Top N Customer Label"]):
        city_to_zip[abt.loc[i, "Customer Name"]] = abt.loc[i, "Top N Customer Label"]
for i in abt.index:
    if pd.isna(abt.loc[i, "Top N Customer Label"]):
        abt.loc[i, "Top N Customer Label"] = city_to_zip.get(abt.loc[i, "Customer Name"])
```

```python

# drop rows where "Total Number of Customer_x", "Total Number of Customer_y","Number of Records" is missing
abt=abt.drop(["Total Number of Customer_x", "Total Number of Customer_y","Number of Records"], axis=1)
abt
```

```python
abt.shape
```

```python
#fill the missing value in the column ship Mode

segment_mapping = abt.dropna(subset=['Ship Mode'])[['Order ID', 'Ship Mode']].drop_duplicates().set_index('Order ID')['Ship Mode'].to_dict()

def fill_segment(row):
    if pd.isna(row['Ship Mode']):
        OrderID = row['Order ID']
        return segment_mapping.get(OrderID, None)  # Fill with corresponding Ship Mode if found, else None
    return row['Ship Mode']

abt['Ship Mode'] = abt.apply(fill_segment, axis=1)
```

```python
#fill the missing value in the column Discount based on mean value

abt["Discount"]=abt["Discount"].fillna(abt["Discount"].mean())
print(abt)
```

```python
#fill the missing value in the column Quantity based on mean value

abt["Quantity"]=abt["Quantity"].fillna(abt["Quantity"].mean())
print(abt)
```

```python
#fill the missing item in the Product Name

abt["Product Name"]=abt["Product Name"].fillna("Not available")
print(abt)
```

```python
abt = abt.T.drop_duplicates().T
print(abt)
```

```python
segment_mapping = abt.dropna(subset=['Segment'])[['Sub-Category', 'Segment']].drop_duplicates().set_index('Sub-Category')['Segment'].to_dict()

# Step 2: Fill missing values in Segment based on the Sub-category
def fill_segment(row):
    if pd.isna(row['Segment']):
        sub_category = row['Sub-Category']
        return segment_mapping.get(sub_category, None)  # Fill with corresponding Segment if found, else None
    return row['Segment']

abt['Segment'] = abt.apply(fill_segment, axis=1)
```

```python
from datetime import datetime
def format_date(date_str):
    for fmt in ('%m/%d/%y', '%Y-%m-%d %H:%M:%S', '%m/%d/%Y'):
        try:
            date_obj = datetime.strptime(date_str, fmt)
            return date_obj.strftime('%m/%d/%y 0:00')
        except ValueError:
            continue
    raise ValueError(f"Unrecognized date format: {date_str}")

# Apply the function to the 'dates' column
abt['Order Date'] = abt['Order Date'].apply(format_date)

print(abt)
```

```python
abt['Order Date'] = pd.to_datetime(abt['Order Date'], errors='coerce')

# Extract the month
abt['month'] = abt['Order Date'].dt.month

print(abt)
```

```python
import math
def custom_round(value):
    return math.floor(value * 100) / 100

# Apply the rounding function to the 'Sales' column
abt['Discount'] = abt['Discount'].apply(custom_round)

print(abt)
```

```python
abt['Total Sales'] = abt['Sales'] / (1 - abt['Discount'])

# Calculate Discount Amount
abt['Discount Amount'] = abt['Total Sales'] - abt['Sales']

# Display the updated DataFrame
print(abt)
```
